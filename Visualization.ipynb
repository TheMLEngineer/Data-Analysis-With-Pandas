## Bar charts

```python
google = data.DataReader(name="GOOG",data_source="yahoo",start='2004-1-1',end='2016-12-31')
```

```python
google.head()
```

```python
def rank_performance(stock_price):
    if stock_price <= 200:
        return "not good"
    elif stock_price>200 and stock_price<=500:
        return "Satisfactory"
    else:
        return "Excellent performance"
```

```python
google['Close'].apply(rank_performance)
```

```python
google['Close'].apply(rank_performance).value_counts()
```

```python
google['Close'].apply(rank_performance).value_counts().plot(kind = 'line')
```

```python
google['Close'].apply(rank_performance).value_counts().plot(kind = 'bar')
```

```python
google['Close'].apply(rank_performance).value_counts().plot(kind = 'barh')
#barh -> horizontal bar
```

## Pie charts

```python
apple = data.DataReader(name='AAPL',data_source="yahoo",start='2012-1-1',end='2016-12-31')
```

```python
apple.head(3)
```

```python
apple['Close'].mean()
```

```python
def rank_performance(stock_price):
    if stock_price >= 93.3617192254726:
        return "Above avg"
    else:
        return "Not above avg"
```

```python
plt.style.use('ggplot')
apple['Close'].apply(rank_performance).value_counts().plot(kind='pie',legend=True)
```

## Histograms

```python
google = data.DataReader(name="GOOG",data_source="yahoo",start='2004-1-1',end='2016-12-31')
```

```python
google.head(3)
```

```python
def custom_round(stock_price):
    return int(stock_price / 100.0) * 100
```

```python
custom_round(149.96)
#Gives value as hundreds
```

```python
google['High'].apply(custom_round)
```

```python
google['High'].apply(custom_round).value_counts()
```

```python
google['High'].apply(custom_round).value_counts().sort_index()
```

```python
google['High'].apply(custom_round).nunique()
```

```python
google['High'].apply(custom_round).plot(kind='hist' , bins = 9)
#Bins value is got by nunique()
```
